Here I am importing relevant packages that will be used.

```python
import numpy as np
import matplotlib as mpl
import pandas as pd
from matplotlib import pyplot as plt
from geopy.geocoders import ArcGIS
nom=ArcGIS()
import folium
```

Using Pandas, I imported each .csv file containing the data and defining each as their own data frame. 

```python
beb = pd.read_csv("BellinghamBay.csv") 
bub = pd.read_csv("BurrowsBay.csv")         
nsd1 = pd.read_csv("NooksackDelta1.csv")
nsd2 = pd.read_csv("NooksackDelta11.csv")
rs1 = pd.read_csv("RosarioStrait1.csv")
rs2 = pd.read_csv("RosarioStrait11.csv")
```

By using folium and ArcGIS from geopy, I made a map centered around the longitude and latitude coordiantes for the City of Bellingham. Then, labeled popup markers (click on marker to see label) of each sample site were added to the map using folium and their longitude and latitiude.

```python
#bellingham coordinates
map=folium.Map(location=[48.5123, -122.4785], zoom_start=9)
map

map.add_child(folium.Marker(location=[48.53622, -122.7078],popup="Rosario Strait Site 2"))
map.add_child(folium.Marker(location=[48.42447, -122.8078],popup="Rosario Strait Site 1"))
map.add_child(folium.Marker(location=[48.47393, -122.6813],popup="Burrows Bay"))
map.add_child(folium.Marker(location=[48.73705, -122.5609],popup="Nooksack Delta Site 1"))
map.add_child(folium.Marker(location=[48.7547, -122.5636],popup="Nooksack Delta Site 2"))
map.add_child(folium.Marker(location=[48.71828, -122.5729],popup="Bellingham Bay"))
```

Here, I am visualizing the temperature at each site with respect to depth. Matplotlib (plt.subplots) was used to creat 6 subplots (one for each site) and the spacing was adjusted (plt.subplots_adjust). I then used Pandas to plot each dataframe (e.g. beb.plot). I also had to set which subplot the dataframes would plot to (e.g. ax=axes[0,0]), the x variable (Temperature), the y variable (Depth), and legend was set to false since one would automatically appear and it is not necessary for these sets of graphs. The y axis had to inverted to show incresing depth rather than decreacing depth. Finally, axis labels and a title were set. The same steps were repeated again for the Salinity and Denisity plots but instead of using Temperature the words Salinity and Density were used to pull that data for their respectice plots.

```python
#temp
fig, axes = plt.subplots(nrows=3, ncols=2)

plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=2,
                    wspace=0.4,
                    hspace=0.4)

beb.plot(ax=axes[0,0], x='Temperature', y='Depth', legend=False)
bub.plot(ax=axes[0,1], x='Temperature ', y='Depth', legend=False)
nsd1.plot(ax=axes[1,0],x='Temperature ', y='Depth', legend=False)
nsd2.plot(ax=axes[1,1], x='Temperature ', y='Depth', legend=False)
rs1.plot(ax=axes[2,0],x='Temperature ', y='Depth', legend=False)
rs2.plot(ax=axes[2,1],x='Temperature ', y='Depth', legend=False)

axes[0,0].invert_yaxis()
axes[0,1].invert_yaxis()
axes[1,0].invert_yaxis()
axes[1,1].invert_yaxis()
axes[2,0].invert_yaxis()
axes[2,1].invert_yaxis()

axes[0,0].set_ylabel('Depth (m)')
axes[0,1].set_ylabel('Depth (m)')
axes[1,0].set_ylabel('Depth (m)')
axes[1,1].set_ylabel('Depth (m)')
axes[2,0].set_ylabel('Depth (m)')
axes[2,1].set_ylabel('Depth (m)')

axes[0,0].set_xlabel('Temperature\N{DEGREE SIGN}C')
axes[0,1].set_xlabel('Temperature\N{DEGREE SIGN}C')
axes[1,0].set_xlabel('Temperature\N{DEGREE SIGN}C')
axes[1,1].set_xlabel('Temperature\N{DEGREE SIGN}C')
axes[2,0].set_xlabel('Temperature\N{DEGREE SIGN}C')
axes[2,1].set_xlabel('Temperature\N{DEGREE SIGN}C')

axes[0,0].set_title('Burrows Bay')
axes[0,1].set_title('Bellingham Bay')
axes[1,0].set_title('Nooksack Delta Site 1')
axes[1,1].set_title('Nooksack Delta Site 2')
axes[2,0].set_title('Rosario Strait Site 1')
axes[2,1].set_title('Rosario Strait Site 2')
```
